```python
import pandas as pd
import glob

# načtení všech CSV souborů jednotlivých krajů ze složky
kraje = glob.glob("C:/Dominika/industrial_analysis/data/mzdy/csv/*.csv")

# spojení do jednoho DataFrame pro celou Českou republiku
mzdy = pd.concat([pd.read_csv(f) for f in kraje], ignore_index=True)

# filtrování napotřebnách řádků
mzdy = mzdy[mzdy["Převažující činnost"] != "Odvětví celkem"].copy()

# upravení názvů Převažující činnosti
mzdy["Převažující činnost"] = mzdy["Převažující činnost"].str[3:]

# převod roků do jednoho sloupce
mzdy = mzdy.melt(
    id_vars=["Kraj", "Převažující činnost"], var_name="Rok", value_name="Mzda_Kc"
)
# převedení na požadovaný datový formát
mzdy["Mzda_Kc"] = (
    mzdy["Mzda_Kc"]
    .replace("\\s", "", regex=True)
    .pipe(pd.to_numeric, errors="coerce")
    .astype("Int64")  
)
mzdy["Rok"] = mzdy["Rok"].astype(int)
mzdy
mzdy.to_csv("mzdy_odvetvi.csv", index=False)
```
