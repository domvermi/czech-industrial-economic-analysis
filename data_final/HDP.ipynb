```python
import pandas as pd

hdp = pd.read_csv("HDP_1990-2025.csv")

# filtrovaní ukazatelů jen na HDP a regionů jen na jednotlivé kraje
hdp = hdp[
    (hdp["Ukazatel"] == "Hrubý domácí produkt (mil. Kč, běžné ceny)")
    & ~hdp["ČR, Reg. soudržnosti, Kraje"].isin([ "Česko", "Moravskoslezsko", "Severozápad", "Jihovýchod", "Střední Morava", "Jihozápad", "Severovýchod", "Střední Čechy"])
].copy()

# odstranění nepotřebných sloupců
hdp = hdp[["ČR, Reg. soudržnosti, Kraje", "Roky", "Hodnota"]]

# přejmenování sloupců a krajů, jejich seřazení
hdp = hdp.rename(
    columns={"ČR, Reg. soudržnosti, Kraje": "Kraj", "Roky": "Rok", "Hodnota": "HDP"}
)
hdp["Kraj"] = hdp["Kraj"].replace({"Hlavní město Praha": "Praha", "Kraj Vysočina" : "Vysočina"})
hdp = hdp.sort_values(["Kraj", "Rok"])

# odstranění a kontrola duplicit
hdp = hdp.drop_duplicates(subset=["Kraj", "Rok", "HDP"])
assert not hdp.duplicated(["Kraj", "Rok"]).any()
hdp = hdp.reset_index(drop=True)
hdp
hdp.to_csv("hdp_1995-2024_final.csv", index=False)
```
