```python
import pandas as pd
dim_rok = pd.read_csv("fact_zamestnanci_odvetvi_1993-2024.csv")
dim_rok = (
    dim_rok[["Rok"]]
    .drop_duplicates()
    .sort_values("Rok")
    .reset_index(drop=True)
)
dim_rok
dim_rok.to_csv("dim_roky.csv", index= False)
```

```python
dim_kraj = pd.read_csv("fact_mzdy_odvetvi.csv")
dim_kraj = (
    dim_kraj[["Kraj"]]
    .drop_duplicates()
    .sort_values("Kraj")
    .reset_index(drop=True)
)
dim_kraj
dim_kraj.to_csv("dim_kraj.csv", index= False)
```

```python
dim_cinnost = pd.read_csv("mzdy_odvetvi.csv")
dim_cinnost = (
    dim_cinnost[["Převažující činnost"]]
    .drop_duplicates()
    .sort_values("Převažující činnost")
    .reset_index(drop=True)
)
dim_cinnost
dim_cinnost.to_csv("dim_cinnost.csv", index= False)
```
